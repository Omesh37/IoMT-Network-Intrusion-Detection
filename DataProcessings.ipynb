```python
import pandas as pd
```

```python
train_df = pd.read_parquet("train.parquet")
test_df = pd.read_parquet("test.parquet")
```

```python
feature_names = train_df.drop(columns=["label"]).columns.tolist()
print(feature_names)
```

```python
for col in feature_names:
    python_name = col.lower().replace(" ", "_")
    print(f'    {python_name}: float = Field(..., alias="{col}")')
```

```python
train_df["canonical_label"] = train_df["label"].str.rstrip("0123456789")
test_df["canonical_label"] = test_df["label"].str.rstrip("0123456789")
```

```python
train_df["canonical_label"].value_counts()
```

```python
category_map = {
    "TCP_IP-DDoS-UDP": "DDoS",
    "TCP_IP-DDoS-ICMP": "DDoS",
    "TCP_IP-DDoS-TCP": "DDoS",
    "TCP_IP-DDoS-SYN": "DDoS",
    "TCP_IP-DoS-UDP": "DoS",
    "TCP_IP-DoS-SYN": "DoS",
    "TCP_IP-DoS-ICMP": "DoS",
    "TCP_IP-DoS-TCP": "DoS",
    "MQTT-DDoS-Connect_Flood": "MQTT",
    "MQTT-DDoS-Publish_Flood": "MQTT",
    "MQTT-DoS-Publish_Flood": "MQTT",
    "MQTT-DoS-Connect_Flood": "MQTT",
    "MQTT-Malformed_Data": "MQTT",
    "Recon-Port_Scan": "Recon",
    "Recon-OS_Scan": "Recon",
    "Recon-VulScan": "Recon",
    "Recon-Ping_Sweep": "Recon",
    "ARP_Spoofing": "Spoofing",
    "Benign": "Benign",
}

train_df["category"] = train_df["canonical_label"].map(category_map)
test_df["category"] = test_df["canonical_label"].map(category_map)
```

```python
train_df["category"].isna().sum()
```

```python
train_df["category"].value_counts()
```

```python
train_df.columns
```

```python
train_df.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(train_df.columns.tolist())
print(test_df.columns.tolist())
```

```python
train_df["binary_label"] = train_df["category"].apply(
    lambda x: "Benign" if x == "Benign" else "Attack"
)

test_df["binary_label"] = test_df["category"].apply(
    lambda x: "Benign" if x == "Benign" else "Attack"
)
```

```python
print(train_df["binary_label"].value_counts())
print(test_df["binary_label"].value_counts())
```

```python
train_df.to_parquet("train_enriched.parquet")
test_df.to_parquet("test_enriched.parquet")
```

```python
non_feature_cols = ["label", "canonical_label", "category", "binary_label"]

X_train = train_df.drop(columns=non_feature_cols)
y_train = train_df["binary_label"]
X_test = test_df.drop(columns=non_feature_cols)
y_test = test_df["binary_label"]
```

```python
rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    class_weight="balanced",
    n_jobs = -1,
    verbose = 2
)

rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
import pandas as pd
importances = pd.Series(rf.feature_importances_, index=X_train.columns).sort_values(ascending=False)
print(importances.head(10))
```

```python
X_test_reset = X_test.reset_index(drop=True)
y_test_reset = y_test.reset_index(drop=True)
y_pred_series = pd.Series(y_pred).reset_index(drop=True)

misclassified = X_test_reset[y_test_reset != y_pred_series]
misclassified_true = y_test_reset[y_test_reset != y_pred_series]

# join back to the original label (not just binary) to see WHICH attack types get missed
test_df_reset = test_df.reset_index(drop=True)
missed_categories = test_df_reset.loc[misclassified.index, "category"].value_counts()
print(missed_categories)
```
